```python
def data(df):
    df = pd.read_csv(df, sep = ";", header = None, dtype = str, encoding = "ISO-8859-1")
    persona = df[15]
    fecha_oper = df[12]
    resultado = list(zip(persona[:], fecha_oper[:]))
    resultado = pd.DataFrame(resultado)
    resultado.columns = ['persona', 'fecha']
    resultado['vol'] = 1
    resultado = resultado.groupby(['fecha','persona'])['vol'].sum().reset_index()
    
    return resultado.pivot(index = 'fecha', columns = 'persona', values = 'vol')
```
